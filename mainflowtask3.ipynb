```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('C:\\Users\\hp\\Downloads\\householdtask3.csv')
```

```python
display(data.head(10))
```

```python
#scatter plot with year against own
plt.scatter(data['year'],data['own'])

#Adding title to the plot
plt.title("Scatter Plot")

#Setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#line chart with year and own
plt.plot(data['year'], data['own'])

#Adding title to the plot
plt.title("Line Chart")

#Setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()

```

```python
#Bar chart or bar plot
plt.bar(data['year'], data['own'])

#Adding title to the plot
plt.title("Bar Chart")

#Setting the x and y label
plt.xlabel('year')
plt.ylabel('own')

#Showing the result
plt.show()
```

```python
#Histogram
plt.hist(data['income'])

plt.title("Histogram")

plt.show()
```
