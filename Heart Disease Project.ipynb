```python
#1 How many people have heart disease and how many people doesn't have heart disease?

df.target.value_counts()
```

```python
df.target.value_counts().plot(kind = 'bar', color = ["black","blue"])
plt.title("Heart Diease Values")
plt.xlabel("1 = Heart Diease, 0 = No Heart Diease")
plt.ylabel("Amount");
```

```python
df.target.value_counts().plot(kind = 'pie' , figsize = (8,6))
plt.legend(["Diease" , "No diease"]);
```

```python
# '0' represent 'Female'
# '1' represent 'Male'
df.sex.value_counts()
```

```python
df.sex.value_counts().plot(kind = 'pie', figsize = (8,6))
plt.legend(["Male", "Female"])
plt.title("Male Female Ratio");
```

```python
#2 People of which sex has most heart diease?

pd.crosstab(df.target , df.sex)
```

```python
sns.countplot(x = 'target' , data = df, hue = 'sex')
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("0 = No Heart Disease, 1 = Heart Disease");
```

```python
#3 people of which sex has which type of sex pain most?

df.cp.value_counts()
```

```python
df.cp.value_counts().plot(kind = 'bar', color = ['salmon', 'green', 'pink', 'orchid'])
plt.title('Chest pain type vs count');
```

```python
pd.crosstab(df.sex , df.cp)
```

```python
pd.crosstab(df.sex, df.cp).plot(kind = 'bar', color = ['skyblue', 'orchid', 'salmon', 'green'])
plt.title('Type of chest pain for sex')
plt.xlabel('0 = Female, 1 = Male');
```

```python
#4 People with which chest pain are most pron to have heart disease?

pd.crosstab(df.cp, df.target)
```
